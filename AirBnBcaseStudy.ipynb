# **Description of Every Column:**



1.   **ID:** The unique id for each hotel/listing in the dataset.
2.   **Name:** The name of every place that appears on AirBnB.
3.   **host_id:** The unique id for each host in our data.
4.   **host_name:** Name of the host that appears on AirBnB.
5.   **neighbourhood_group** This is a grouping of neighbours that appears in the same city.
6.   **neighbourhood:** Specific neighbourhood where the place is located.
7.   **latitude:** Geographical Latitude of the hotel.
8.   **longitude:** Geographical longitude of the hotel/place.
9.   **room_type:** This is the type of room offered by the place, such as Hotel, Private room , apartment, shared room.
10.  **price:** Stay charges per night in US dollars.
11.  **minimum_nights:** For how long the guests must stay at the place/hotel.
12.  **number_of_reviews:** Total reviews we have for the particular place.
13.  **last_review:** The most recent review (latest review) for the place.
14.  **review_per_month:** How many reviews we have for that place in a month.
15.  **calculated_host_listing_count:** This is the total number of listings that the host has on AirBnB.
16.  **availability_365:** This is the number of days in the next 365 days that the hotel is available for booking.

```python
plt.boxplot(df['price'])
plt.title('Price')
plt.show()
```

```python
# To Remove Outliers:
# Step 1: Calculate the Q1 and Q3
# Step 2: Calculate IQR = Q3-Q1
# Step 3: Calulcate lower_limit and upper_limit
# Step 4: Select the rows that are greater than or equal to lower limit and the rows that are smaller
# than or equal to upper limit.

Q1 = df.price.quantile(0.25)
Q3 = df.price.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)
df = df[(df.price>=lower_limit) & (df.price<=upper_limit)]
```
